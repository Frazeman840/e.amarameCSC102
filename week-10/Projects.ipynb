```python
import math
class cylinder():
    radius = 0
    height = 0
    volume=0
    
    def __init__(self,r,h):
        self.radius= r
        self.height= h
        
    def display(self):
        print("Volume of cylinder = " + str(self.volume))
        
    def calculate(self):
        self.volume = math.pi * (self.radius ** 2) * self.height
        num1 = int(input("Radius: "))
        num2 = int(input("Height: "))
        
    
class cube():
    lenght=0
    area=0
    
    def __init__(self,l):
        self.lenght = l
        
    def display(self):
        print("Area of Cube = " + str(self.area))
        
    def calculate(self):
        self.area = 6 * (self.lenght ** 2)
        num1 = int(input("Lenght: "))
        

class parallelogram():
    base = 0
    altitude=0
    area=0
    
    def __init__(self,b,a):
        self.base = b
        self.altitude = a
        
    def display(self):
        print("Area of Parallelogram = " + str(self.area))
        
    def calculate(self):
        self.area = self.base * self.altitude
        num1 = int(input("Base: "))
        num2 = int(input("Altitude: "))
        
class rhombus():
    diagonal1 = 0
    diagonal2 = 0
    area= 0
    
    def __init__(self,d1,d2):
        self.diagonal1 = d1
        self.diagonal2 = d2
        
    def display(self):
        #print("Diagonal one = " + str(self.diagonal1))
        #print("Diagonal two = " + str(self.diagonal2))
        print("Area of Rhombus = " + str(self.area))
        
    def calculate(self):
        self.area = 1/2 * self.diagonal1 * self.diagonal2
        num1 = int(input("Diagonal1: "))
        num2 = int(input("Diagonal2: "))

class trapezuim():
    height=0
    base1= 0
    base2= 0
    area=0
    
    def __init__(self,h,b1,b2):
        self.height = h
        self.base1 = b1
        self.base2 = b2
        
    def display(self):
        #print("Height = " + str(self.height))
        #print("base1 = " + str(self.base1))
        #print("base2 = " + str(self.base2))
        print("Area of Trapezuim = " + str(self.area))
        
    def calculate(self):
        num1 = int(input("Height: "))
        num2 = int(input("base1: "))
        num3 = int(input("base2: "))
        self.area = self.height/2 * (self.base1 + self.base2)



print("Select Equation: Press 1 for Area of a Trapezuim")
print("                 Press 2 for Area of a Rhombus")
print("                 Press 3 for Area of a Parallelogram")
print("                 Press 4 for Area of a Cube")
print("                 Press 5 for Volume of a Cylinder")

Equation = int(input("Number for Equation: "))

if Equation == 1:
    obj = trapezuim(num1,num2,num3)
    obj.calculate()
    obj.display()
elif Equation == 2:
    obj = rhombus(num1,num2)
    obj.calculate()
    obj.display()
elif Equation == 3:
    obj = parallelogram(num1,num2)
    obj.calculate()
    obj.display()
elif Equation == 4:
    obj = cube(num1)
    obj.calculate()
    obj.display()
elif Equation == 5:
    obj = cylinder(num1,num2)
    obj.calculate()
    obj.display()
    
```
